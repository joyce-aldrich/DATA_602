```
# Aggregate data by 'Geo Place Name' and find mean pollutant values
top5_locations = df_cleaned.groupby('Geo Place Name')['Data Value'].mean().nlargest(5)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(top5_locations.index, top5_locations.values, color='blue')
plt.title("Top 5 Locations with Highest Average Pollutant Levels", fontsize=14)
plt.xlabel("Location", fontsize=12)
plt.ylabel("Average Pollutant Levels", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

Observation:
Among the five boroughs of New York City, the top 5 locations with the highest average pollutant levels are all located in the Bronx.

```
# Extract year from 'Start_Date'
df_cleaned['Year'] = pd.to_datetime(df_cleaned['Start_Date']).dt.year

# Aggregate by year
yearly_trends = df_cleaned.groupby('Year')['Data Value'].mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_trends.index, yearly_trends.values, marker='o', linestyle='-', color='red')
plt.title("Yearly Trend in Average Pollutant Levels", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Pollutant Levels", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Oberservation:
After 2005, the average pollutant levels have shown a consistent decreasing trend over time, with a significant drop observed during the COVID-19 pandemic period, from the beginning of 2020 to 2022

```
# Bar plot using Seaborn
sns.barplot(x=top5_locations.index, y=top5_locations.values)
plt.title("Top 5 Locations with Highest Average Pollutant Levels", fontsize=14)
plt.xlabel("Location", fontsize=12)
plt.ylabel("Average Pollutant Levels", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Line plot using Seaborn
sns.lineplot(x=yearly_trends.index, y=yearly_trends.values, marker='o', color='blue')
plt.title("Yearly Trend in Average Pollutant Levels", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Pollutant Levels", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Overall:
Matplotlib: Offers flexibility and granular control over every aspect of the plot. However, it requires more code to produce polished visuals.

Seaborn: Built on top of Matplotlib, it simplifies common tasks and produces aesthetically pleasing visualizations with less code. It's better for statistical plots and comes with preset themes.
